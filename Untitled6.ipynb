#Ejercicio 4.5) Remplazar los nulos que aparecen en la variable "thal por 999. *No* guardarlo en el dataframe original. Usar fillna.


```python
import pandas as pd

# Leer CSV
url = "https://raw.githubusercontent.com/fpineyro/homework-0/master/heart_disease.csv"
df = pd.read_csv(url)

# Reemplazar NaN en thal por 999, sin modificar el original
df_thal_reemplazo = df["thal"].fillna(999)

# Mostrar resultado
print(df_thal_reemplazo)
```

#Ejercicio 4.6) Remplazar los nulos que aparecen en la variable "thal" por 999 (como antes),
# y también remplazar los nulos de "num_vessels_flour" por el promedio de la variable.
# Guardar el resultado en un dataframe nuevo llamado "d_hd2".

```python
import pandas as pd

# Leer CSV
url = "https://raw.githubusercontent.com/fpineyro/homework-0/master/heart_disease.csv"
df = pd.read_csv(url)

# Crear copia para no tocar el original
d_hd2 = df.copy()

# 1. Reemplazar NaN en 'thal' por 999
d_hd2["thal"] = d_hd2["thal"].fillna(999)

# 2. Reemplazar NaN en 'num_vessels_flour' por el promedio
mean_num_vessels = d_hd2["num_vessels_flour"].mean(skipna=True)
d_hd2["num_vessels_flour"] = d_hd2["num_vessels_flour"].fillna(mean_num_vessels)

# Mostrar resultado
print(d_hd2.head())
```

#Ejercicio 4.7) Chequear resultado con status

```python
# Chequear si quedaron NaN en las variables de interés
print("Valores nulos en d_hd2:")
print(d_hd2.isna().sum())

# Otra forma de ver el estado del DataFrame
print("\nResumen de d_hd2:")
print(d_hd2.info())
```

#Ejercicio 4.8)
# Eliminar todas los registros con nulos usando `dropna` de pandas

```python
import pandas as pd

# Leer CSV original
url = "https://raw.githubusercontent.com/fpineyro/homework-0/master/heart_disease.csv"
df = pd.read_csv(url)

# Eliminar todos los registros con nulos
df_sin_nulos = df.dropna()

# Verificar resultado
print(df_sin_nulos.info())
print("Cantidad de registros originales:", len(df))
print("Cantidad de registros sin nulos:", len(df_sin_nulos))
```
